```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/sudheerachary/rainfall-prediction/master/data/rainfall_in_india_1901-2015.csv")

df1 = pd.DataFrame(data[['SUBDIVISION','YEAR', 'JAN', 'FEB', 'MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC', 'ANNUAL']])

df1[0:100]
```

```python
df_new = df2[df2.columns[1:15]]
df_new.describe()
```

# Linear Regression


```python

from sklearn.model_selection import train_test_split
df_new1= df_new.dropna() #dropping nan values
X=np.array((df_new1['YEAR']))
Y=np.array((df_new1['ANNUAL']))
xTrain, xTest, yTrain, yTest = train_test_split(X,Y, test_size = 0.2)
print(X)
print("shape of original dataset :", df_new1.shape)
print("shape of input - training set", xTrain.shape)
print("shape of output - training set", yTrain.shape)
print("shape of input - testing set", xTest.shape)
print("shape of output - testing set", yTest.shape)
```

```python
import numpy as np


x = np.array(df_new1['YEAR'])
x = x.reshape(-1,1)
print(x.shape)

y = np.array(df_new1['ANNUAL'])
y = y.reshape(-1,1)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression().fit(X.reshape(-1,1),Y.reshape(-1,1))
```

```python
y_predict = model.predict(xTest.reshape(-1,1))
print(y_predict)
```

# Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(xTrain, yTrain, color='red')
b, m = model.intercept_, model.coef_[-1]
plt.plot(X, m*X + b)
```
